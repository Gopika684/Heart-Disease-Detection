```python
X=df.drop(columns='target',axis=1)
```

```python
Y=df['target']
```

**SPLITTING DATASET INTO TRAIN AND TEST**

```python
from sklearn.model_selection import train_test_split # Importing train_test_split from sklearn.model_selection

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=23)
```

```python
print(X_test.shape,X_train.shape)
```

# MODEL TRAINING

```python
from sklearn.linear_model import LogisticRegression # Importing LogisticRegression from sklearn.linear_model

model = LogisticRegression()
```

```python
#Train the model with Training data
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
Y_train
```

```python
from sklearn.metrics import accuracy_score # Importing the accuracy_score function

Y_pred=model.predict(X_test)

score = accuracy_score(Y_test,Y_pred)
```

```python
print("The accuracy score is",score)
```
